```python
import pandas as pd
import os

df_list = []

for file in os.listdir():
    if file.endswith(".csv"):
        print("Loading:", file)   # To confirm files are being read
        
        temp_df = pd.read_csv(file)
        
        # Extract city name from file name
        city_name = file.split("_")[0]
        temp_df["City"] = city_name
        
        df_list.append(temp_df)

print("\nNumber of files loaded:", len(df_list))

combined_df = pd.concat(df_list, ignore_index=True)

print("\nFinal Combined Shape:", combined_df.shape)

combined_df.head()
```

```python
combined_df = combined_df.loc[:, ~combined_df.columns.str.contains('^Unnamed')]

combined_df.info()
```

```python
combined_df["Date"] = pd.to_datetime(combined_df["Date"], dayfirst=True)

combined_df["Year"] = combined_df["Date"].dt.year
combined_df["Month"] = combined_df["Date"].dt.month
```

```python
combined_df["Date"] = pd.to_datetime(
    combined_df["Date"],
    format="%d/%m/%y"
)
```

```python
combined_df = combined_df.dropna(subset=["AQI"])
```

### City-wise AQI Comparison

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average AQI by city
city_avg = combined_df.groupby("City")["AQI"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))

# Create barplot
sns.barplot(
    x=city_avg.index,
    y=city_avg.values,
    palette="Reds_r"
)

# Add value labels on top of bars
for index, value in enumerate(city_avg.values):
    plt.text(index, value + 2, f"{value:.1f}", 
             ha='center', fontsize=10)

plt.title("Average AQI Comparison Across Major Indian Cities", fontsize=14)
plt.xlabel("City")
plt.ylabel("Average AQI")
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

####  Delhi shows the highest average AQI among selected cities.
####  Southern cities like Bangalore and Hyderabad show comparatively lower AQI.
####  There is a clear disparity in air quality levels across metro regions.

### Which City Has the Highest Average AQI?

```python
# Calculate average AQI by city
city_avg_aqi = combined_df.groupby("City")["AQI"].mean().sort_values(ascending=False)

# Display top city
highest_city = city_avg_aqi.idxmax()
highest_value = city_avg_aqi.max()

print("City with Highest Average AQI:", highest_city)
print("Average AQI:", round(highest_value, 2))

city_avg_aqi.head()
```

##### Delhi has the highest average AQI (208.29), indicating significantly worse air quality compared to Mumbai (106.78), Hyderabad (85.40), Chennai (77.34), and Bangalore (74.32), which show comparatively lower pollution levels.

### Which City Improved Over Years?

```python
import numpy as np

city_trend = {}

for city in combined_df["City"].unique():
    city_data = combined_df[combined_df["City"] == city]
    yearly_avg = city_data.groupby("Year")["AQI"].mean()
    
    # Fit linear trend line
    slope = np.polyfit(yearly_avg.index, yearly_avg.values, 1)[0]
    city_trend[city] = slope

trend_df = pd.Series(city_trend).sort_values()

print("Cities with Improvement (Negative Slope = AQI Decreasing):")
trend_df
```

### Which Year Had Worst Pollution?

```python
# Average AQI per year
year_avg_aqi = combined_df.groupby("Year")["AQI"].mean().sort_values(ascending=False)

worst_year = year_avg_aqi.idxmax()
worst_value = year_avg_aqi.max()

print("Year with Worst Pollution:", worst_year)
print("Average AQI:", round(worst_value, 2))

year_avg_aqi.head()
```

##### The analysis indicates that 2018 recorded the highest average AQI (125.71), making it the most polluted year in the dataset, followed by 2019 and 2023.

### Is AQI Increasing Trend-wise Overall?

```python
# Overall yearly average AQI
overall_yearly = combined_df.groupby("Year")["AQI"].mean()

# Calculate slope
overall_slope = np.polyfit(overall_yearly.index, overall_yearly.values, 1)[0]

if overall_slope > 0:
    print("AQI is Increasing Trend-wise")
elif overall_slope < 0:
    print("AQI is Decreasing Trend-wise")
else:
    print("AQI is Stable")

print("Trend Slope Value:", round(overall_slope, 4))
```

##### AQI is Decreasing Trend-wise 📉
#### Trend Slope Value: -2.7993
